# Clean up data using pandas

For this section of the project, I am cleaning mock data with pandas.

```python
import pandas as pd
```

```python
patients = pd.read_csv('vet_data.txt', header=None, sep='//', engine='python')
```

```python
df = pd.DataFrame(patients)
df
```

```python
df = df.replace(regex=r';;', value=',')
df
```

```python
row = []

for field in df:
    row.append(df[field].tolist())

print(row)
```

```python
patients = pd.DataFrame(row)

patients
```

```python
patients = patients[0].str.split(',', expand=True)

patients
```

```python
headers = patients.iloc[0]
patients = patients[1:]
patients.columns = headers

patients
```

```python
patients.columns
```

```python
patients.columns = patients.columns.str.strip()
```

```python
patients.columns
```

```python
patients = patients.rename({'Pet name': 'pet', 'owner name': 'owner', 'pet type': 'type', 'pet dob': 'dob'}, axis=1)
```
